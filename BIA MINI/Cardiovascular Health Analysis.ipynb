```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from graphviz import Digraph

# Load the dataset
data = pd.read_csv('C:/Users/DELL/Desktop/heart.csv')  # Replace 'your_dataset.csv' with your data file.

# Basic data exploration
print(data.head())
print(data.describe())

# 1. Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.xlabel('Age')
plt.title('Age Distribution')
plt.show()

```

```python
# 2. Bar plot of Chest Pain Type
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='ChestPainType')
plt.xlabel('Chest Pain Type')
plt.title('Chest Pain Type Distribution')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

```python
plt.figure(figsize=(6, 6))
data['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Oldpeak'], bins=20, kde=True)
plt.xlabel('Oldpeak')
plt.title('Oldpeak Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='FastingBS')
plt.xlabel('Fasting Blood Sugar (0: No, 1: Yes)')
plt.title('Fasting Blood Sugar Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Age', y='MaxHR', ci=None)
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.title('Max Heart Rate Over Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='ChestPainType', hue='HeartDisease')
plt.xlabel('Chest Pain Type')
plt.title('Heart Disease Presence by Chest Pain Type')
plt.legend(title='Heart Disease', labels=['Absent', 'Present'])
plt.show()
```
